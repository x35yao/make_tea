# Combine the analyze result of different object

```python
from evaluate import combine_h5files
from utils import get_h5files
import os
vid_ids = ['1645721497', '1645722954', '1645724115', '1646154227', '1646155040']
filtertype = 'median'
objs = ['teabag','cup', 'pitcher', 'tap']
cameras = ['left']

combined_files = []
for vid_id in vid_ids:
    h5files = get_h5files(vid_id, objs, filtertype, cameras)
    combined_file = combine_h5files(h5files)
    combined_files.append(combined_file)
    
```

# Visualizing the combined file

```python
from utils import get_videos
from visualization import create_video_with_h5file
videos = []

for vid_id in vid_ids:
    video = get_videos(vid_id, cameras = ['left'])
    videos += video

for video, h5file in zip(videos, combined_files):
    create_video_with_h5file(video, h5file)
```
