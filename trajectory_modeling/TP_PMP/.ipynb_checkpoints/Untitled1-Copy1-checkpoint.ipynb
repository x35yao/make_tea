```python
import os
import pandas as pd
import numpy as np
'''
Convert the axis so that it is easier to understand:
x' = z
y' = y
z' = -x
'''
def load_gripper_data(base_dir, action_ind=0):
    demos = os.listdir(base_dir)
    data = {}
    for demo in demos:
        demo_dir = os.path.join(base_dir, demo)
        root, dirs, files = next(os.walk(demo_dir))
        if len(dirs) == 0:
            continue
        for f in files:
            if 'Servo' in f:
                servo_file = os.path.join(demo_dir, f)
            elif 'NDI' in f:
                ndi_file = os.path.join(demo_dir, f)
        # Open servo file to find the start and end of an action
        with open(servo_file, 'r') as f:
            lines = f.readlines()
            t_start = float(lines[1 + action_ind].split(',')[0])
            t_end = float(lines[2 + action_ind].split(',')[0])
        # Open NDI file and extract the parts corresponding to the queried action
        df_temp1 = pd.read_csv(ndi_file)
        duration = t_end - t_start
        idx_start = df_temp1['Time'].sub(t_start).abs().idxmin()
        idx_end = df_temp1['Time'].sub(t_end).abs().idxmin()
        df_ndi = df_temp1[idx_start: idx_end]
        df_ndi['Duration'] = df_ndi['Time'] - df_ndi['Time'].iloc[0]

        # traj_obj_cam.append(df_cam)
        data[demo] = df_ndi[['x', 'y', 'z', 'Duration']]
    return data

def load_markers_data(base_dir, triangulation = 'leastereo', action_ind=0):
    demos = os.listdir(base_dir)
    data = {}
    for demo in demos:
        demo_dir = os.path.join(base_dir, demo)
        root, dirs, files = next(os.walk(demo_dir))
        if len(dirs) == 0:
            continue
        for f in files:
            if 'Servo' in f:
                servo_file = os.path.join(demo_dir, f)
            elif 'NDI' in f:
                ndi_file = os.path.join(demo_dir, f)
        if triangulation == 'leastereo':
            obj_markers_file = os.path.join(demo_dir, 'leastereo', 'markers_trajectory_3d.h5')
        elif triangulation == 'dlc3d':
            obj_markers_file = os.path.join(demo_dir, 'dlc3d', 'markers_trajectory_3d.h5')
        # Open servo file to find the start and end of an action
        with open(servo_file, 'r') as f:
            lines = f.readlines()
            t_start = float(lines[1 + action_ind].split(',')[0])
            t_end = float(lines[2 + action_ind].split(',')[0])
        df_temp1 = pd.read_csv(ndi_file)
        idx_start = df_temp1['Time'].sub(t_start).abs().idxmin()
        idx_end = df_temp1['Time'].sub(t_end).abs().idxmin()
        duration = t_end - t_start
        df_temp2 = pd.read_hdf(obj_markers_file)
        idx_start_cam = int(len(df_temp2) * idx_start / len(df_temp1))
        idx_end_cam = int(len(df_temp2) * idx_end / len(df_temp1))
        df_cam = df_temp2[idx_start_cam: idx_end_cam].reset_index(drop=True)
        df_cam['Duration'] = np.arange(len(df_cam))/len(df_cam)*duration
        data[demo] = df_cam
    return data

base_dir = '/home/luke/Desktop/project/Process_data/postprocessed/2022-05-26'
gripper_trajectorys = load_gripper_data(base_dir, action_ind = 0)
markers_trajectorys_lea = load_markers_data(base_dir, triangulation = 'leastereo', action_ind = 0)
markers_trajectorys_dlc = load_markers_data(base_dir, triangulation = 'dlc3d', action_ind = 0)
```

```python
def homogenous_transform(R,vect):

    '''
    :param R: 3x3 matrix
    :param vect: list x,y,z
    :return:Homogenous transformation 4x4 matrix using R and vect
    '''

    H = np.zeros((4,4))
    H[0:3,0:3] = R
    if not isinstance(vect, list):
        frame_displacement = list(vect) + [1]
    else:
        frame_displacement = vect + [1]
    D = np.array(frame_displacement)
    D.shape = (1,4)
    H[:,3] = D
    return H

def inverse_homogenous_transform(H):

    '''
    :param H: Homogenous Transform Matrix
    :return: Inverse Homegenous Transform Matrix
    '''


    R = H[0:3,0:3]
    origin = H[:-1,3]
    origin.shape = (3,1)

    R = R.T
    origin = -R.dot(origin)
    return homogenous_transform(R,list(origin.flatten()))

def homogenous_position(vect):
    temp = np.array(vect)
    ones = np.ones(temp.shape[0])
    return np.c_[temp, ones]
```

```python
from scipy.spatial.transform import Rotation as R

def get_initial_positions(df, window_size = 50):
    df_partial = df.iloc[:window_size]
    df_median = df_partial.median()
    return df_median
    

def prototype_to_obj(prototype, df):
    idx = pd.IndexSlice
    bodyparts = df.index.get_level_values('bodyparts').unique()
    points_prototype = []
    points_obj = []
    for bodypart in bodyparts:
        df_bodypart = df.loc[idx[:,:,bodypart]]
        if not df_bodypart.isnull().values.any():
            points_prototype.append(prototype[bodypart])
            points_obj.append(df_bodypart.to_numpy())
            
    vectors_prototype = []
    vectors_obj =[]
    for i, (prototype_point,obj_point) in enumerate(zip(points_prototype, points_obj)):
        for j,(other_prototype_point,other_obj_point)in enumerate(zip(points_prototype[i+1:], points_obj[i+1:])):
            vectors_prototype.append(prototype_point - other_prototype_point)
            vectors_obj.append(obj_point - other_obj_point)
    r, error = R.align_vectors(vectors_obj, vectors_prototype)
    rotmatrix = r.as_matrix()
    translations = np.array(points_obj) - (rotmatrix @ np.array(points_prototype).T).T
    translation = np.median(translations, axis = 0)
    
    H = homogenous_transform(rotmatrix, translation)
    points_prototype_transformed = (H @ homogenous_position(points_prototype).T).T
    dists = []
    for i, point in enumerate(points_prototype_transformed):
        point = point[:3]
        print(f'point_transfered:{point}')
        print(f'point of the object {points_obj[i]}')
        dists.append(np.linalg.norm(point[:3] - points_obj[i]))
    print(f'average distance: {np.mean(np.array(dists), axis = 0)}')
    return H
```
